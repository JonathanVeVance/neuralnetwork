```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
df_x = pd.DataFrame(iris.data, columns = iris.feature_names)
df_y = pd.DataFrame(iris.target)
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.33, random_state=42)
```

```python
def misclass_error(ytrue, ypred):
    misclassifications = np.sum(ytrue != ypred)
    return misclassifications / len(ytrue)
```

```python
x_train_np = x_train.to_numpy()
y_train_np = y_train.to_numpy()
x_test_np = x_test.to_numpy()
y_test_np = y_test.to_numpy()
```

```python
from neuralnetwork import network

NN = network()
NN.add_layer(10, 'relu')
NN.add_layer(5, 'relu')
NN.add_layer(1, 'sigmoid')

from helpers.accelerator import adam, momentum, rmsprop
from helpers.regularizer import l2_regularizer
# accel = adam(0.9, 0.9)
# accel = momentum(0.5)
accel = rmsprop(0.9)
reg = l2_regularizer(1e2)
```

```python
NN.train_network(x_train_np, y_train_np, learning_rate = 1e-5, grad_descent_type = 'minibatchgd',
                loss_function = 'crossentropy', regularizer = reg, accelerator = accel, batch_size = 30)

ypreds = NN.predict(x_test_np)
ypreds = np.where(ypreds > 0.5, 1, 0)
print(misclass_error(y_test_np, ypreds))
```
